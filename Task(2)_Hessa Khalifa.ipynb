# Tasks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Task (2) Dataset.csv')
```

```python
df.head()
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
Columns= df[['JamsCount','TrafficIndexLive','JamsDelay','JamsLength','TrafficIndexWeekAgo','TravelTimeHistoric','TravelTimeLive']]
corrorelation = Columns.corr()
sns.heatmap(corrorelation, annot=True)
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data= df, x='JamsCount', y='TrafficIndexWeekAgo')
plt.title('Relationship Between JamsCount and TrafficIndexWeekAgo')
plt.xlabel('JamsCount')
plt.ylabel('TrafficIndexWeekAgo')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data= df, x='JamsDelay', y='TrafficIndexWeekAgo')
plt.title('Relationship Between JamsDelay and TrafficIndexWeekAgo')
plt.xlabel('JamsDelay')
plt.ylabel('TrafficIndexWeekAgo')
plt.show()
```
